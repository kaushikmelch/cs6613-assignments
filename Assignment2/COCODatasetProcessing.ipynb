```python
from pycocotools.coco import COCO
import os
import shutil

annFile = 'annotations/instances_train2017.json'
coco = COCO(annFile)

classes_to_include = ['potted plant', 'giraffe', 'traffic light', 'fire hydrant', 'bottle', 'parking meter', 'boat', 'kite', 'chair', 'car', 'train', 'keyboard', 'person', 'umbrella', 'stop sign', 'bird', 'bowl', 'tv', 'bus', 'laptop', 'clock', 'airplane', 'bench', 'cup']  # Replace with your desired classes

catIds = coco.getCatIds(catNms=classes_to_include)
```

```python
catIds = coco.getCatIds(catNms=classes_to_include)
```

```python
catIds
```

```python
imgIds = []
for x in catIds:
    imgIds.append(coco.getImgIds(catIds = [x]))
```

```python
imgIds
```

```python
finalIds = []
for x in imgIds:
    count = 0;
    for y in x:
        if(count < 200):
            finalIds.append(y)
        count += 1
```

```python
len(finalIds)
```

```python
coco.download(imgIds = finalIds, tarDir = 'filtered_ds')
```
